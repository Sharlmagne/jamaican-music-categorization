```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('tracks_classified_3_continuous.csv')
```

```python
df.head()
```

#### Drop Columns

```python
df = df.drop(['artist_id','track_id','time_signature'], axis=1)
```

```python
for column in df:
    df.rename(columns={column:column.capitalize()}, inplace=True)
    
df.rename(columns={'Artist_name':'Artist Name', 'Track_name':'Track Name', 'Release_year':'Release Year'}, inplace=True)
```

```python
df.head()
```

#### Drop Duplicates

```python
df.duplicated().sum()
```

```python
df2 = df.drop_duplicates(subset=['Track Name', 'Clusters'], 
                   keep='first')
```

```python
df2.duplicated().sum()
```

#### Round Tempo Values

```python
df2['Tempo'] = round(df2['Tempo'])
```

```python
df2.head()
```

#### Normalize Loudness Column

```python
#Normalize the Loudness Column
from sklearn.preprocessing import MinMaxScaler
import numpy as np

df3 = df2.copy()
df3['Loudness'] = MinMaxScaler().fit_transform(np.array(df3['Loudness']).reshape(-1,1))
```

```python
df3.head()
```

#### Create Clusters

```python
clusters = list(df3['Clusters'])
```

```python
for i in range(len(clusters)):
    clusters[i]= f'Cluster {clusters[i]}'
```

```python
df4 = df3.copy()
df4['Clusters'] = clusters
```

```python
df4.head()
```

```python
df4.to_csv('tracks_powerbi.csv', index=False)
```
