```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('Files/tracks_classified.csv')
```

```python
df.head()
```

#### Drop Columns

```python
df = df.drop(['artist_id','track_id','time_signature'], axis=1)
```

```python
for column in df:
    df.rename(columns={column:column.capitalize()}, inplace=True)
    
df.rename(columns={'Artist_name':'Artist Name', 'Track_name':'Track Name', 'Release_year':'Release Year'}, inplace=True)
```

```python
df.head()
```

#### Drop Duplicates

```python
df.duplicated().sum()
```

```python
df2 = df.drop_duplicates(subset=['Track Name', 'Clusters'], 
                   keep='first')
```

```python
df2.duplicated().sum()
```

```python
df2.head()
```

#### Normalize Loudness Column

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Normalize the Loudness Column
df2['Loudness'] = MinMaxScaler().fit_transform(np.array(df2['Loudness']).reshape(-1,1))
```

#### Create Clusters

```python
df2['Clusters'] = df2['Clusters'] + 1
```

```python
clusters = list(df2['Clusters'])
```

```python
for i in range(len(clusters)):
    clusters[i]= f'Cluster {clusters[i]}'
```

```python
df3 = df2.copy()
df3['Clusters'] = clusters
```

```python
df3.head()
```

#### Change Categorical Columns

```python
df4 = df3.copy()
```

```python
# Edit categotical values to reflect what they represent.
notes = ['C','C#/Db','D','D#/Eb','E','F','F#/Gb','G','G#/Ab','A','A#/Bb','B']
for i in range(12):
    df4['Key'] = df4['Key'].replace([i],notes[i])

df4['Mode'] = df4['Mode'].replace([1],'Major')
df4['Mode'] = df4['Mode'].replace([0],'Minor')

df4['Explicit'] = df4['Explicit'].replace([True],'Explicit')
df4['Explicit'] = df4['Explicit'].replace([False],'Clean')
```

```python
df4.head()
```

```python
df4.to_csv('Files/tracks_final.csv', index=False)
```
